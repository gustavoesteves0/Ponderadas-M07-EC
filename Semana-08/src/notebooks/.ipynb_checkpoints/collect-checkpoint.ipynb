```python
import pandas as pd
import matplotlib.pyplot as plt
import requests
```

```python
# Pegando os dados da Render Token pela API do CoinGecko
def get_crypto_data(crypto_id, vs_currency, days):
    url = f'https://api.coingecko.com/api/v3/coins/{crypto_id}/market_chart'
    params = {'vs_currency': vs_currency, 'days': days}
    response = requests.get(url, params=params)
    data = response.json()
    return data
```

```python
# Função para processar e organizar os dados em um DataFrame
def process_data(data):
    prices = data['prices']
    df = pd.DataFrame(prices, columns=['timestamp', 'price'])
    
    # Verifica se os dados de volume estão disponíveis
    if 'total_volumes' in data:
        volumes = data['total_volumes']
        df['volume'] = [vol[1] for vol in volumes]
    else:
        df['volume'] = None  # Ou pode-se optar por omitir essa coluna
    
    df['date'] = pd.to_datetime(df['timestamp'], unit='ms')
    df.set_index('date', inplace=True)
    return df[['price', 'volume']]
```

```python
# Parametros
crypto_id = 'render-token'
vs_currency = 'usd'
days = 365  # Fetching data for one year
```

```python
# Buscar e processar os dados
data = get_crypto_data(crypto_id, vs_currency, days)
crypto_df = process_data(data)
```

```python
# Verifica se o índice de datas possui uma frequência associada
crypto_df = crypto_df.asfreq('D')  # 'D' define a frequência diária (Day)
```

```python
# Exibe as primeiras linhas do DF
print(crypto_df.head())
```

```python
# Plotando os preços ao longo do tempo
plt.figure(figsize=(10, 6))
plt.plot(crypto_df.index, crypto_df['price'], label='Preço')
plt.title('Variação do Preço do Render Token')
plt.xlabel('Data')
plt.ylabel('Preço (USD)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Plotando o volume de negociação ao longo do tempo
plt.figure(figsize=(10, 6))
plt.plot(crypto_df.index, crypto_df['volume'], color='orange', label='Volume de Negociação')
plt.title('Volume de Negociação do Render Token')
plt.xlabel('Data')
plt.ylabel('Volume')
plt.grid(True)
plt.legend()
plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
# Definindo o modelo ARIMA
model = ARIMA(crypto_df['price'], order=(5, 1, 0))  # Ajuste os parâmetros de acordo com a série
model_fit = model.fit()
```

```python
# Fazendo previsões para os próximos 30 dias
forecast = model_fit.forecast(steps=30)
```

```python
# Exibindo previsões
print(forecast)
```
